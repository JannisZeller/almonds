## Plotting the C++ gernerated data
-----

```python
## Imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
## Reding the data
data = pd.read_csv("./data/data.csv", sep=",").astype(np.float32)
X = data.values

## Checking the shape
print(f"Plotting a Mandelbrot set with {X.shape[1]} x {X.shape[0]} pixels.")
plt.imshow(X, cmap="hot")
plt.axis("off")

## Saving the image 
plt.savefig(f"./img/{datetime.now().strftime('%d-%m-%Y_%H-%M-%S')}.jpg", bbox_inches="tight", dpi=1000)
```
